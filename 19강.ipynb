```python
titanic=pd.read_csv("data/titanic.csv")
```

```python
titanic["age"]=titanic["age"].replace(["?"],[None]).astype(float)
```

```python
titanic
```

```python
df=titanic[["sex","pclass","survived","age"]]
```

```python
gbo=df.groupby(by="sex")
```

```python
gbo
```

```python
gbo.get_group("female")[["pclass","survived","sex","age"]]
```

```python
titanic[titanic["sex"]=="male"][["pclass","survived","age"]].mean()
```

```python
titanic.groupby("sex").agg({"age":["max","min"],"pclass":"mean"})
```

```python
def range(a):
    return a.max()-a.min()
```

```python
titanic.groupby("pclass")["age"].agg(["max","min",range])
```
