```python
펭귄의 부리길이가 늘어나수록 팔길이도 같이 늘어난다
지역마다 펭귄의 부리와 팔길이의 분포비율이가 다르다
Adelie 와 Gentoo는 펭귄의 수가 많아 선이 겹겹이 싸여있지만 Chinstrap은 수가 적어 선이 적다
세가지 펭귄 종 중에서 Gentoo가 부리길이와 날개길이 관계가 가장 정비례에 가깝다. 즉, 부리가 길 때, 날개가 길다고 표현하기 가장 적절한 종이다. 
Gentoo의 경우, 부리와 날개의 길이가 짧을 수록 상관관계가 높은 것이 밀집되어 있고, 
 Chinstrap의 경우 부리와 날개의 길이가 길수록 상관관계가 높은 것이 밀집되어 있다. 
Adelie의 경우, 전반적인 펭귄의 부리길이 날개 길이가 서로대강 연관되어있다.
```

```python
tip=sns.load_dataset("tips")
```

```python
tip
```

```python
sns.rugplot(data=tip,y="tip",height=0.2)
```

```python
sns.kdeplot(data=tip,x="total_bill")
sns.rugplot(data=tip,x="total_bill",height=0.1)
```

```python
sns.rugplot(data=tip,x="total_bill",y="tip",height=0.025)
sns.scatterplot(data=tip,x="total_bill",y="tip")
# rug의 total_bill의 선이 모여있는곳과 tip선이 모여있는 
# 곳에scatterplot의 원이 가장 많이 모여있다
# total_bill이 올라갈수록 tip도 정비례 하게 올라간다
```

```python
sns.displot(data=penguins,x="body_mass_g",hue="sex",col="species")
# 각 종 마다 펭귄의 무게를 와 수를 구하였다 
# 남자와 여자를 색갈로 구별하여 겹치는 부분은 색깔이 다르다
# adelie와Chinstrap은 몸무게가 작은 펭귄들이 서식하지만 Gentoo는 
# 몸무게가 많이 나가는 펭귄들이 서식하고있다
# adelie와 Gentoo는 펭귄의 수가 많지만 chinstrap의 펭귄 수는 다른 
# 종에 비해 수가 적다
```

```python
sns.displot(data=tip,x="tip",col="time",kind="kde",rug=True)
lunch에 tip이 더 적고 dinner의 tip이 더 많다
tip이 rug 선이 몰려 있으면 kde의 그래프가 올라간다
```
