```python
penguins=sns.load_dataset("penguins")
```

```python
penguins
```

```python
sns.histplot(data=penguins,x="body_mass_g",hue="species",multiple="stack",bins=20)
```

```python
sns.histplot(data=tips, x="tip", kde=True)
```

```python
sns.set_theme()
plt.figure(dpi=200)
sns.histplot(data=penguins,x="body_mass_g",hue="species",multiple="stack",bins=20,kde=True)
plt.title("Penguin Body Mass")
```

```python

sns.set_theme()
plt.figure(dpi=200)   
sns.histplot(
    data=penguins, 
    x="body_mass_g", 
    bins=20, 
    hue="species", 
    multiple="stack", 
    kde=True
)
plt.title("Penguin Body Mass")
```

```python

sns.set_theme()
plt.figure(dpi=200)   
sns.distplot(
    data=penguins, 
    x="body_mass_g", 
    bins=20, 
    hue="species", 
    multiple="stack", 
    kde=True
)
plt.title("Penguin Body Mass")
```
