```python
died=[10,25,5,30,21]
germinated=[74,88,56,69,59]
plt.figure(figsize=(6,6))
z=['Spinach','Turnip','Rhubarb','Broccoli','Kale']
plt.bar(z,died,color='b',width=0.7)
plt.bar(z,germinated,color='g',width=0.4,align='edge')
plt.ylabel('Number Of Dead Seedling')
plt.title('Failure Rates')
```

```python
died=[10,25,5,30,21]
germinated=[74,88,56,69,59]
plt.figure(figsize=(6,6))
z=['Spinach','Turnip','Rhubarb','Broccoli','Kale']
plt.bar(z,died,color='b',bottom=germinated,label='Failed Seedlings')
plt.bar(z,germinated,color='g',label="Germinated Seedlings")
plt.title('Greenhouse Seedlings')
plt.legend()
```

```python
first_class=titanic[titanic["pclass"]==1]["age"]
second_class=titanic[titanic["pclass"]==2]["age"]
third_class=titanic[titanic["pclass"]==3]["age"]
plt.figure(figsize=(10,4))
ax=plt.subplot(1,3,1)
plt.hist(first_class,label='1st Class',color="yellow",bins=50)
plt.title("1st Class")
ax=plt.subplot(1,3,2,sharey=ax)
plt.hist(second_class,label='2st Class',color="b",bins=50)
plt.title("2st Class")
ax=plt.subplot(1,3,3,sharey=ax)
plt.hist(third_class,label='3st Class',color="red",bins=50)
plt.title("3st Class")
# Bins      . X  . 
```
