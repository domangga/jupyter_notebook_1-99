```python
import pandas as pd
```

```python
from wordcloud import WordCloud
from string import digits
import re
```

```python
text=""
with open("data/대화.txt","r",encoding="utf-8")as f:
    lines=f.readlines()
for line in lines:
    text+= line.replace("어","").replace("응","").replace("사진\n","").replace("ㅋ","").replace("삭제된 메시지입니다.\n","").replace("ㅎ","").replace("준원","").replace("양민석","").replace("오후","").replace("준형","")
    text = re.sub(r'[0-9]+', '', text)
print(text)


```

```python
font_path = 'C:/Windows/Fonts/malgun.ttf'
wc = WordCloud(font_path=font_path, background_color="white", width=600, height=400)
wc.generate(text)
wc.to_file("a.png")
```

```python
import matplotlib.pyplot as plt
from string import digits
from PIL import Image
import numpy as np

font_path = 'C:/Windows/Fonts/malgun.ttf'
wc = WordCloud(font_path=font_path, background_color="white", width=600, height=400)
wc.generate(text)
wc.to_file("result_review.png")
plt.imshow(wc)
plt.title("Word Frequency", size = 13)
plt.axis('off')
plt.show()
```

```python
im=Image.open('d.png')
mask_arr = np.array(im) 


wc = WordCloud(font_path=font_path, background_color="yellow", mask = mask_arr, prefer_horizontal = True, width=600, height=400)
wc.generate(text)
wc.to_file("result_review.png")
plt.imshow(wc)
plt.title("Word Frequency", size = 13)
plt.axis('off')
plt.show()
```
