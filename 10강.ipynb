```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x, y,'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
```

```python
plt.subplot(1,4,1)
plt.plot(x,y,'y--')
plt.subplot(1,4,4)
plt.plot(y,x,'b*-');
# Plt에 matplotlib 을 저장했고, matplotlib에 있는 plot이라는 메소드를 통해 어떤 
# 상관관계가 있는 그래프를 그릴 것인지를 지정하고, 해당 그래프의 색깔을 지정한다. 
```

```python
fig=plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x, y,'b'),axes2.plot(y,x,'r')
axes1.set_xlabel('x_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes2.set_xlabel('x_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')
axes2.set_title('Axes 2 Title')

# Figure 라는 메소드는 그래프의 크기와 가로 세로 비율을 정할수 있도록 해준다
# fig.add_axes([왼쪽축에서의 거리, 바닥에서의 거리, 그래프의 폭,그래프의 높이])
```

```python
fig=plt.figure()
axes=fig.add_axes([1,1,1.5,1])
axes.plot(x, x+1,"midnightblue",linewidth=0.1)
axes.plot(x, x+2,"navy",linewidth=0.5)
axes.plot(x, x+3,"darkblue",linewidth=1)
axes.plot(x, x+4,"b",linewidth=2)
```

```python
fig,ax = plt.subplots(figsize=(12,6))
ax.plot(x, x+1,"midnightblue",linewidth=0.1)
ax.plot(x, x+2,"navy",linewidth=0.5)
ax.plot(x, x+3,"darkblue",linewidth=1)
ax.plot(x, x+4,"b",linewidth=2)
ax.plot(x, x+5,"g",linewidth=2.5)
ax.plot(x, x+6,"g",linewidth=3,ls="-.")
ax.plot(x, x+7,"g",linewidth=3.5,ls=":")

ax.plot(x, x+9,"b",linewidth=3,ls="-",marker='+')
ax.plot(x, x+10,"b",linewidth=3,ls="--",marker='o')
ax.plot(x, x+11,"b",linewidth=3,ls="-",marker='s')
ax.plot(x, x+12,"b",linewidth=3,ls="--",marker='1')

# Figsize = 그래프의 사이즈를 가로 * 세로 비율로 확정하는 것. 
```
